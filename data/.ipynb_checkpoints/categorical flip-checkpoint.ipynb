```python
#load data
#Import libraries
import numpy as np
import pandas as pd 

import pyomo.environ as pyo
from pyomo.opt import SolverFactory

#load data
xls = pd.ExcelFile('data\case6.xlsx')
# xls = pd.ExcelFile('data\case14.xlsx')
# xls = pd.ExcelFile('data\IEEE118_.xlsx')
wind_load = pd.read_csv('data\IEEE118_Load_Wind.csv')
branch = pd.read_excel(xls, 'branch')
gen = pd.read_excel(xls, 'gen')
bus = pd.read_excel(xls, 'bus')
cost = pd.read_excel(xls, 'cost')

con_cost = cost['cons']

#Y matrix
base = 100
bus_len = bus.shape[0]
Ybus = np.zeros([bus_len,bus_len])
gen_len = len(gen['bus'].ravel())
x = [pd.DataFrame() for i in range(bus_len)]

for i in range(bus_len):  
    x[i] = branch.loc[branch['fbus'] == i]
    fro = [j for j in x[i]['tbus']]
    fro = pd.DataFrame(fro)
    for j in range(len(fro)):
        Ybus[i-1,fro.loc[j][0]-1] =  -1/pd.DataFrame(x[i]['x']).iloc[j][0]
        Ybus[fro.loc[j][0]-1,i-1] =  -1/pd.DataFrame(x[i]['x']).iloc[j][0]

for i in range(bus_len):
    Ybus[i,i] = np.sum(Ybus[i,:])*(-1)


# declare variables
Pd = np.zeros(bus_len)
Qd = np.zeros(bus_len)
Vmax = np.zeros(bus_len)
Vmin = np.zeros(bus_len)
Pmax = np.zeros(bus_len)
Pmin = np.zeros(bus_len)

for i in range(bus_len):
    Pd[i] = bus['Pd'][i]/base
    Qd[i] = bus['Qd'][i]/base
    Vmax[i] = bus['Vmax'][i]
    Vmin[i] = bus['Vmin'][i]

for i in range(gen_len):
    k = i
    i = gen['bus'].ravel()[i]
    Pmax[i-1] = gen.loc[gen['bus'] == i]['Pmax'].ravel()[0]/base
    Pmin[i-1] = gen.loc[gen['bus'] == i]['Pmin'].ravel()[0]/base


p = pd.DataFrame([gen['Pmin'],gen['Pmax']]).T
p.columns = ['minimum', 'maximum']
p = p.T
p.columns = ['P%s' % (i+1) for i in range(gen_len)]
p = p.T/base

fva = round(p,2)
fva
```

```python
#first n random selections
def _generate_sample(aims):
    
    #Create a model
    model = ConcreteModel()
    model.t = Var(range(bus_len))
    model.v = Var(range(bus_len))
    model.vars = Var(range(len(fva)))
    model.pd = Var(range(len(fva)))
    

    #Constraints
    model.c = ConstraintList()
    model.c.add(model.t[0] == 0)  
    for i in range(len(fva)):
        model.c.add(expr = model.vars[i] <= fva.iloc[i]['maximum'])
        model.c.add(expr = model.vars[i] >= fva.iloc[i]['minimum'])
        model.c.add(expr = model.vars[i] == np.dot(aims[i],source[i]))

        #load constraint
        model.c.add(expr = model.pd[i] <= Phigh)
        model.c.add(expr = model.pd[i] >= Plow)
    for i in range(bus_len):
        model.c.add(expr = model.v[i] <= Vmax[i])
        model.c.add(expr = model.v[i] >= Vmin[i]) 
        
  

    #Nodal equations
    count = 0
    for i in range(bus_len):
        if (np.any(gen['bus'].ravel()-1==[i])):
            holder1 = model.vars[count]
            holder2 = model.pd[count]
            count = count+1
        else:
            holder1 = 0
            holder2 = 0
        model.c.add(expr = sum([Ybus[i,j]*model.t[j] for j in range(bus_len)]) - holder1 + holder2 == 0)

    #Line flow constraints
    for i in range(len(branch)): 
        x = int(branch.loc[i]['fbus'])-1
        y = int(branch.loc[i]['tbus'])-1
        val = (branch.loc[i]['rateA']/100)
        model.c.add(expr = (Ybus[x,y])* (model.t[x]-model.t[y]) <= int(branch.loc[i]['rateA'])/100)
        model.c.add(expr = (Ybus[x,y])* (model.t[y]-model.t[x]) <= int(branch.loc[i]['rateA'])/100)


    solution = SolverFactory('ipopt').solve(model)
   
    if (solution.solver.status == SolverStatus.ok) and (solution.solver.termination_condition == TerminationCondition.optimal):     
        result = [abs(round(model.vars[i].value,2)) for i in range(len(fva))]

        return result

    else:
        return None
    
#Script
import time
import numpy as np
import pandas as pd
import bisect
import random
import cmath
import itertools
from pyomo.environ import *
from pyomo.opt import SolverFactory

#Load bounds for optimisation
Plow = min(Pd)
Phigh = max(Pd)*4

#define step for discretization
size = 4; ideal = 1/size;
#define discrete step for each generator
source = [list(np.linspace(fva.iloc[i]['minimum'],fva.iloc[i]['maximum'],size)) for i in range(len(fva))]
source = [[round(k,2) for i,k in enumerate(source[i])] for i in range(len(source))]

#All possible combinations (discrete levels)
a_combinations = list(map(list,itertools.product([0, 1], repeat=size)))
#Relevant combinations
t_combinations = []
for i in range(len(a_combinations)):
    if sum(a_combinations[i]) == 1:
        t_combinations.append(a_combinations[i])

#define list to track discrete-samples generated (binary)
tracker = []
#define list to track actual samples generated 
samples = []
##define list to track unique targets proposed
aims = []
#number of samples
n = 3; true_s = 0; infeasible_count = 0;


for i in range(n):
    while(1):
        trial = random.choices(t_combinations,k=len(fva))
        if trial not in aims:
            aims.append(trial)
            new_sample =  _generate_sample(aims[i])

            if new_sample is not None:
                tracker.append(aims[i])
                samples.append(new_sample)
                true_s += 1
            else:
                infeasible_count += 1        
            break
```

```python
#Further sampling
def _generate_sample(targets, weights):
    
    #Create a model
    model = ConcreteModel()
    model.t = Var(range(bus_len))
    model.v = Var(range(bus_len))
    model.vars = Var(range(len(fva)))
    model.pd = Var(range(len(fva)))

    model.x = Var(range(size),range(len(fva)), within=Binary, bounds=(0,1)) 
    

    #Constraints
    model.c = ConstraintList()
    model.c.add(model.t[0] == 0)
    
    #actual Gen MW
    for i in range(len(fva)):
        model.c.add(expr = sum(model.x[j,i] for j in range(size)) == 1)
        model.c.add(expr = model.vars[i] == sum(model.x[j,i]*source[i][j] for j in range(size)))
                
    
    quad_exp = 0
    for i in range(len(fva)):
        #load constraint
        model.c.add(expr = model.pd[i] <= Phigh)
        model.c.add(expr = model.pd[i] >= Plow)
        #Obj function
        quad_exp += weights[i] * (model.vars[i]-targets[i])**2    #weights[i] *

    count = 0
    for i in range(bus_len):
        #Bus voltage constraints
        model.c.add(expr = model.v[i] <= Vmax[i])
        model.c.add(expr = model.v[i] >= Vmin[i]) 
        
        #Nodal equations
        if (np.any(gen['bus'].ravel()-1==[i])):
            holder1 = model.vars[count]
            holder2 = model.pd[count]
            count = count+1
        else:
            holder1 = 0
            holder2 = 0
        model.c.add(expr = sum([Ybus[i,j]*model.t[j] for j in range(bus_len)]) - holder1 + holder2 == 0)

    #Line flow constraints
    for i in range(len(branch)): 
        x = int(branch.loc[i]['fbus'])-1
        y = int(branch.loc[i]['tbus'])-1
        val = (branch.loc[i]['rateA']/100)
        model.c.add(expr = (Ybus[x,y])* (model.t[x]-model.t[y]) <= int(branch.loc[i]['rateA'])/100)
        model.c.add(expr = (Ybus[x,y])* (model.t[y]-model.t[x]) <= int(branch.loc[i]['rateA'])/100)

    model.cost = Objective(expr = quad_exp)


    solution = SolverFactory('mindtpy').solve(model, mip_solver='glpk', nlp_solver='ipopt')
   
    if (solution.solver.status == SolverStatus.ok) and (solution.solver.termination_condition == TerminationCondition.optimal):
        result = [abs(round(model.vars[i].value,2)) for i in range(len(fva))]
        
#         loads = [round(model.pd[i].value,2) for i in range(len(fva))]
        sequence = [[int(model.x[j,i].value) for j in range(size)] for i in range(len(fva))]

        return result, sequence #, loads

    else:
        return None, None

gen_samples = size**len(fva)
# gen_samples = 64

#begin loop
max_tries = None; try_ = 0;
#Define weight parameters
w_0 = 100
base_val = 10

while True:
    if max_tries is not None and try_ >= max_tries:
        break
    try_ += 1    
    
    #Update weights
    w = sum(np.array(tracker))/len(tracker)
    weights = w_0*abs(w-ideal) + base_val
    print(w)
    
    #Define targets
    #rank weights based on distance to the convergence point
#     ranking = (-abs(w-ideal)).argsort()

    #figure out which part of the plane they fall into
    temp = ((w-ideal)<= 0).astype(int)
    
    temp = ((w-ideal)< 0).astype(int)
    for i in range(len(fva)):
        for j in range(size):
            if w[i][j] == ideal:
                temp[i][j] = round(random.random())    
    targets = temp
    
    
    #Introduce randomness somehow???

    #Make sure only one generator is selected
    min_values = [np.argmin((w-ideal)[i]) for i in range(len(fva))]
    

    for k in range(len(fva)):
        if(sum(targets[k]) > 1):
            max_index = np.argmin(w[k]-ideal)
            non_zero = list(np.nonzero(targets[k])[0])
            non_zero.remove(max_index)
            t = random.sample(non_zero,len(non_zero))
            targets[k,max_index] = 1
            targets[k,t] = 0
        if(sum(targets[k]) < 1):
#             targets[k][min_values[k]] = 1
            targets[k][random.sample([i for i in range(size)],1)] = 1
    
    if [list(targets[i]) for i in range(len(fva))] in tracker:
        ranking = [(-abs(w[i]-ideal)).argsort()[-1] for i in range(len(fva))]
        for i in range(len(fva)):
            targets[i][ranking[i]] = (targets[i][ranking[i]]+1)%2   
    
    print(targets)

    
    weights = [sum((weights*targets)[i]) for i in range(len(fva))]
    print(weights)

    #Get actual targets
    a_targets = [np.dot(targets[i],source[i]) for i in range(len(fva))]


    #Generate new sample
    new_sample, r_target=  _generate_sample(a_targets, weights)
    
    if new_sample is not None:
        print(r_target)
        print("\n")
        tracker.append(r_target)
        samples.append(new_sample)
        true_s += 1
    else:
        infeasible_count += 1 
        print("Infeasible")
        tracker.append(r_target)
    if len(samples) >= gen_samples: break
```
