```python
#Run time?
#load data
#Import libraries
import numpy as np
import pandas as pd 

import pyomo.environ as pyo
from pyomo.opt import SolverFactory

#load data
# xls = pd.ExcelFile('data\case6.xlsx')
# xls = pd.ExcelFile('data\case14.xlsx')
xls = pd.ExcelFile('data\case39.xlsx')
# xls = pd.ExcelFile('data\case68.xlsx')
# xls = pd.ExcelFile('data\IEEE118_.xlsx')
wind_load = pd.read_csv('data\IEEE118_Load_Wind.csv')
branch = pd.read_excel(xls, 'branch')
gen = pd.read_excel(xls, 'gen')
bus = pd.read_excel(xls, 'bus')
cost = pd.read_excel(xls, 'cost')

con_cost = cost['cons']

#Y matrix
base = 100
bus_len = bus.shape[0]
Ybus = np.zeros([bus_len,bus_len])
gen_len = len(gen['bus'].ravel())
x = [pd.DataFrame() for i in range(bus_len)]

for i in range(bus_len):  
    x[i] = branch.loc[branch['fbus'] == i]
    fro = [j for j in x[i]['tbus']]
    fro = pd.DataFrame(fro)
    for j in range(len(fro)):
        Ybus[i-1,fro.loc[j][0]-1] =  -1/pd.DataFrame(x[i]['x']).iloc[j][0]
        Ybus[fro.loc[j][0]-1,i-1] =  -1/pd.DataFrame(x[i]['x']).iloc[j][0]

for i in range(bus_len):
    Ybus[i,i] = np.sum(Ybus[i,:])*(-1)


# declare variables
Pd = np.zeros(bus_len)
Qd = np.zeros(bus_len)
Vmax = np.zeros(bus_len)
Vmin = np.zeros(bus_len)
Pmax = np.zeros(bus_len)
Pmin = np.zeros(bus_len)
Qmax = np.zeros(bus_len)
Qmin = np.zeros(bus_len)

for i in range(bus_len):
    Pd[i] = bus['Pd'][i]/base
    Qd[i] = bus['Qd'][i]/base
    Vmax[i] = bus['Vmax'][i]
    Vmin[i] = bus['Vmin'][i]

for i in range(gen_len):
    k = i
    i = gen['bus'].ravel()[i]
    Pmax[i-1] = gen.loc[gen['bus'] == i]['Pmax'].ravel()[0]/base
    Pmin[i-1] = gen.loc[gen['bus'] == i]['Pmin'].ravel()[0]/base
    Qmax[i-1] = gen.loc[gen['bus'] == i]['Qmax'].ravel()[0]/base
    Qmin[i-1] = gen.loc[gen['bus'] == i]['Qmin'].ravel()[0]/base

p = pd.DataFrame([gen['Pmin'],gen['Pmax']]).T
p.columns = ['minimum', 'maximum']
p = p.T
p.columns = ['P%s' % (i+1) for i in range(gen_len)]
p = p.T/base

q = pd.DataFrame([gen['Qmin'],gen['Qmax']]).T
q.columns = ['minimum', 'maximum']
q = q.T
q.columns = ['Q%s' % (i+1) for i in range(gen_len)]
q = q.T/base

fva = round(p,2)
# fva = pd.concat([round(p,2),round(q,2)])
fva
```

```python
## All brute force
def _generate_sample(aims):
    
    #Create a model
    model = ConcreteModel()
    model.t = Var(range(bus_len))
    model.v = Var(range(bus_len))
    model.vars = Var(range(len(fva)))
    model.qg = Var(range(gen_len))
    model.pd = Var(range(bus_len))
    model.qd = Var(range(bus_len))
    

    #Constraints
    model.c = ConstraintList()
    model.c.add(model.t[0] == 0)  
    for i in range(len(fva)):
        model.c.add(expr = model.vars[i] <= fva.iloc[i]['maximum'])
        model.c.add(expr = model.vars[i] >= fva.iloc[i]['minimum'])
        model.c.add(expr = model.qg[i] <= Qmax[i])
        model.c.add(expr = model.qg[i] >= Qmin[i])
        model.c.add(expr = model.vars[i] == source[i][aims[i]])

        #load constraint
        model.c.add(expr = model.pd[i] <= Phigh)
        model.c.add(expr = model.pd[i] >= Plow)
        model.c.add(expr = model.qd[i] <= Qhigh)
        model.c.add(expr = model.qd[i] >= Qlow)
        

    #Nodal equations
    count = 0
    for i in range(bus_len):
        if (np.any(gen['bus'].ravel()-1==[i])):
            holder1 = model.vars[count]
#             holder2 = model.vars[count+gen_len]
            holder2 = model.qg[count]
            count = count+1
        else:
            holder1 = 0
            holder2 = 0
#         model.c.add(expr = sum([Ybus[i,j]*model.t[j] for j in range(bus_len)]) - holder1 + holder2 == 0)
        model.c.add(expr = sum([model.v[i]*model.v[j]*cmath.polar(Ybus[i,j])[0]*cos(model.t[i]-model.t[j]-cmath.polar(Ybus[i,j])[1]) for j in range(bus_len)]) - holder1 + model.pd[i] == 0)
        model.c.add(expr = sum([model.v[i]*model.v[j]*cmath.polar(Ybus[i,j])[0]*sin(model.t[i]-model.t[j]-cmath.polar(Ybus[i,j])[1]) for j in range(bus_len)]) - holder2 + model.qd[i] == 0)

        # Voltage constraints
        model.c.add(expr = model.v[i] <= Vmax[i])
        model.c.add(expr = model.v[i] >= Vmin[i])

    #Line flow constraints
    for i in range(len(branch)): 
        x = int(branch.loc[i]['fbus'])-1
        y = int(branch.loc[i]['tbus'])-1
        val = (branch.loc[i]['rateA']/100)
#         model.c.add(expr = (Ybus[x,y])* (model.t[x]-model.t[y]) <= int(branch.loc[i]['rateA'])/100)
#         model.c.add(expr = (Ybus[x,y])* (model.t[y]-model.t[x]) <= int(branch.loc[i]['rateA'])/100)
        Pxy = (model.v[x]**2)*cmath.polar(Ybus[x,y])[0]*cos(cmath.polar(Ybus[x,y])[1]) - model.v[x]*model.v[y]*cmath.polar(Ybus[x,y])[0]*cos(model.t[x]-model.t[y]-cmath.polar(Ybus[x,y])[1])
        Qxy =-(model.v[x]**2)*cmath.polar(Ybus[x,y])[0]*sin(cmath.polar(Ybus[x,y])[1]) - model.v[x]*model.v[y]*cmath.polar(Ybus[x,y])[0]*sin(model.t[x]-model.t[y]-cmath.polar(Ybus[x,y])[1])     

        Pyx = (model.v[y]**2)*cmath.polar(Ybus[x,y])[0]*cos(cmath.polar(Ybus[x,y])[1]) - model.v[x]*model.v[y]*cmath.polar(Ybus[x,y])[0]*cos(model.t[y]-model.t[x]-cmath.polar(Ybus[x,y])[1])
        Qyx =-(model.v[y]**2)*cmath.polar(Ybus[x,y])[0]*sin(cmath.polar(Ybus[x,y])[1]) - model.v[x]*model.v[y]*cmath.polar(Ybus[x,y])[0]*sin(model.t[y]-model.t[x]-cmath.polar(Ybus[x,y])[1])     

        model.c.add(expr = (Pxy**2 + Qxy**2)<= val**2 )
        model.c.add(expr = (Pyx**2 + Qyx**2)<= val**2 )


    solution = SolverFactory('ipopt').solve(model)
   
    if (solution.solver.status == SolverStatus.ok) and (solution.solver.termination_condition == TerminationCondition.optimal):     
        result = [abs(round(model.vars[i].value,2)) for i in range(len(fva))]

        return result

    else:
        return None
    
#Script
import time
import numpy as np
import pandas as pd
import bisect
import random
import cmath
import itertools

from pyomo.environ import *
from pyomo.opt import SolverFactory
#Load bounds for optimisation
Plow = min(Pd)
Phigh = max(Pd)*4
Qlow = min(Qd)#*-20
Qhigh = max(Qd)#*20

#define step for discretization
size = 2; ideal = 1/size;
#define discrete step for each generator
source = [list(np.linspace(fva.iloc[i]['minimum'],fva.iloc[i]['maximum'],size)) for i in range(len(fva))]
source = [[round(k,2) for i,k in enumerate(source[i])] for i in range(len(source))]



#number of iterations
iterations = 10
time_k = []
for k in range(iterations):
    #All Combinations
    a_combinations = list(map(list,itertools.product([0, 1], repeat=len(fva))))
    #define list to track discrete-samples generated
    tracker = []
    #keeps track of unique samples generated
    samples = []
    #number of samples
    n = 2**len(fva); true_s = 0; infeasible_count = 0;    
    
    s_time = time.time()
    for i in range(n):
        t = a_combinations.pop()
        new_sample =  _generate_sample(t)
        if new_sample is not None:
            tracker.append(t)
            samples.append(new_sample)
#             a_combinations.remove(a_combinations[i])
            true_s += 1
        else:
            infeasible_count += 1

    time_k.append(time.time() - s_time) 
```

```python
## Proposed approach

#first n random selections
def _generate_sample_1(aims):
    
    #Create a model
    model = ConcreteModel()
    model.t = Var(range(bus_len))
    model.v = Var(range(bus_len))
    model.vars = Var(range(len(fva)))
    model.qg = Var(range(gen_len))
    model.pd = Var(range(bus_len))
    model.qd = Var(range(bus_len))
    

    #Constraints
    model.c = ConstraintList()
    model.c.add(model.t[0] == 0)  
    for i in range(len(fva)):
        model.c.add(expr = model.vars[i] <= fva.iloc[i]['maximum'])
        model.c.add(expr = model.vars[i] >= fva.iloc[i]['minimum'])
        model.c.add(expr = model.qg[i] <= Qmax[i])
        model.c.add(expr = model.qg[i] >= Qmin[i])
        model.c.add(expr = model.vars[i] == source[i][aims[i]])

        #load constraint
        model.c.add(expr = model.pd[i] <= Phigh)
        model.c.add(expr = model.pd[i] >= Plow)
        model.c.add(expr = model.qd[i] <= Qhigh)
        model.c.add(expr = model.qd[i] >= Qlow)
        

    #Nodal equations
    count = 0
    for i in range(bus_len):
        if (np.any(gen['bus'].ravel()-1==[i])):
            holder1 = model.vars[count]
            holder2 = model.qg[count]
            count = count+1
        else:
            holder1 = 0
            holder2 = 0
        model.c.add(expr = sum([model.v[i]*model.v[j]*cmath.polar(Ybus[i,j])[0]*cos(model.t[i]-model.t[j]-cmath.polar(Ybus[i,j])[1]) for j in range(bus_len)]) - holder1 + model.pd[i] == 0)
        model.c.add(expr = sum([model.v[i]*model.v[j]*cmath.polar(Ybus[i,j])[0]*sin(model.t[i]-model.t[j]-cmath.polar(Ybus[i,j])[1]) for j in range(bus_len)]) - holder2 + model.qd[i] == 0)

        # Voltage constraints
        model.c.add(expr = model.v[i] <= Vmax[i])
        model.c.add(expr = model.v[i] >= Vmin[i])

    #Line flow constraints
    for i in range(len(branch)): 
        x = int(branch.loc[i]['fbus'])-1
        y = int(branch.loc[i]['tbus'])-1
        val = (branch.loc[i]['rateA']/100)

        Pxy = (model.v[x]**2)*cmath.polar(Ybus[x,y])[0]*cos(cmath.polar(Ybus[x,y])[1]) - model.v[x]*model.v[y]*cmath.polar(Ybus[x,y])[0]*cos(model.t[x]-model.t[y]-cmath.polar(Ybus[x,y])[1])
        Qxy =-(model.v[x]**2)*cmath.polar(Ybus[x,y])[0]*sin(cmath.polar(Ybus[x,y])[1]) - model.v[x]*model.v[y]*cmath.polar(Ybus[x,y])[0]*sin(model.t[x]-model.t[y]-cmath.polar(Ybus[x,y])[1])     

        Pyx = (model.v[y]**2)*cmath.polar(Ybus[x,y])[0]*cos(cmath.polar(Ybus[x,y])[1]) - model.v[x]*model.v[y]*cmath.polar(Ybus[x,y])[0]*cos(model.t[y]-model.t[x]-cmath.polar(Ybus[x,y])[1])
        Qyx =-(model.v[y]**2)*cmath.polar(Ybus[x,y])[0]*sin(cmath.polar(Ybus[x,y])[1]) - model.v[x]*model.v[y]*cmath.polar(Ybus[x,y])[0]*sin(model.t[y]-model.t[x]-cmath.polar(Ybus[x,y])[1])     

        model.c.add(expr = (Pxy**2 + Qxy**2)<= val**2 )
        model.c.add(expr = (Pyx**2 + Qyx**2)<= val**2 )


    solution = SolverFactory('ipopt').solve(model)
   
    if (solution.solver.status == SolverStatus.ok) and (solution.solver.termination_condition == TerminationCondition.optimal):     
        result = [abs(round(model.vars[i].value,2)) for i in range(len(fva))]

        return result

    else:
        return None
    

#Subsequent sampling
def _generate_sample_2(targets, weights):
    
    #Create a model
    model = ConcreteModel()
    model.t = Var(range(bus_len))
    model.v = Var(range(bus_len))
    model.vars = Var(range(len(fva)))
    model.qg = Var(range(gen_len))
    model.pd = Var(range(bus_len))
    model.qd = Var(range(bus_len))

    model.x = Var(range(size),range(len(fva)), within=Binary, bounds=(0,1)) 
    

    #Constraints
    model.c = ConstraintList()
    model.c.add(model.t[0] == 0)
    
    #actual Gen MW
    for i in range(len(fva)):
        model.c.add(expr = sum(model.x[j,i] for j in range(size)) == 1)
        model.c.add(expr = model.vars[i] == sum(model.x[j,i]*source[i][j] for j in range(size)))
                
    
    quad_exp = 0
    for i in range(len(fva)):
        #load constraint
        model.c.add(expr = model.pd[i] <= Phigh)
        model.c.add(expr = model.pd[i] >= Plow)
        model.c.add(expr = model.qd[i] <= Qhigh)
        model.c.add(expr = model.qd[i] >= Qlow)
        #Obj function
        quad_exp += weights[i] * (model.vars[i]-targets[i])**2   

    count = 0
    for i in range(bus_len):
        #Bus voltage constraints
        model.c.add(expr = model.v[i] <= Vmax[i])
        model.c.add(expr = model.v[i] >= Vmin[i]) 
        
        #Nodal equations
        if (np.any(gen['bus'].ravel()-1==[i])):
            holder1 = model.vars[count]
            holder2 = model.qg[count]

            count = count+1
        else:
            holder1 = 0
            holder2 = 0
        model.c.add(expr = sum([model.v[i]*model.v[j]*cmath.polar(Ybus[i,j])[0]*cos(model.t[i]-model.t[j]-cmath.polar(Ybus[i,j])[1]) for j in range(bus_len)]) - holder1 + model.pd[i] == 0)
        model.c.add(expr = sum([model.v[i]*model.v[j]*cmath.polar(Ybus[i,j])[0]*sin(model.t[i]-model.t[j]-cmath.polar(Ybus[i,j])[1]) for j in range(bus_len)]) - holder2 + model.qd[i] == 0)


    #Line flow constraints
    for i in range(len(branch)): 
        x = int(branch.loc[i]['fbus'])-1
        y = int(branch.loc[i]['tbus'])-1
        val = (branch.loc[i]['rateA']/100)
        Pxy = (model.v[x]**2)*cmath.polar(Ybus[x,y])[0]*cos(cmath.polar(Ybus[x,y])[1]) - model.v[x]*model.v[y]*cmath.polar(Ybus[x,y])[0]*cos(model.t[x]-model.t[y]-cmath.polar(Ybus[x,y])[1])
        Qxy =-(model.v[x]**2)*cmath.polar(Ybus[x,y])[0]*sin(cmath.polar(Ybus[x,y])[1]) - model.v[x]*model.v[y]*cmath.polar(Ybus[x,y])[0]*sin(model.t[x]-model.t[y]-cmath.polar(Ybus[x,y])[1])     

        Pyx = (model.v[y]**2)*cmath.polar(Ybus[x,y])[0]*cos(cmath.polar(Ybus[x,y])[1]) - model.v[x]*model.v[y]*cmath.polar(Ybus[x,y])[0]*cos(model.t[y]-model.t[x]-cmath.polar(Ybus[x,y])[1])
        Qyx =-(model.v[y]**2)*cmath.polar(Ybus[x,y])[0]*sin(cmath.polar(Ybus[x,y])[1]) - model.v[x]*model.v[y]*cmath.polar(Ybus[x,y])[0]*sin(model.t[y]-model.t[x]-cmath.polar(Ybus[x,y])[1])     

        model.c.add(expr = (Pxy**2 + Qxy**2)<= val**2 )
        model.c.add(expr = (Pyx**2 + Qyx**2)<= val**2 )
    
    model.cost = Objective(expr = quad_exp)

    solution = SolverFactory('ipopt').solve(model)
   
    if (solution.solver.status == SolverStatus.ok) and (solution.solver.termination_condition == TerminationCondition.optimal):
        result = [abs(round(model.vars[i].value,2)) for i in range(len(fva))]
        
#         loads = [round(model.pd[i].value,2) for i in range(len(fva))]
#         [model.x[j,i].value for j in range(size) for i in range(len(fva))]

        return result#, loads

    else:
        return None        
    
    
#Script
import time
import numpy as np
import pandas as pd
import bisect
import random
import cmath
import itertools

from pyomo.environ import *
from pyomo.opt import SolverFactory
#Load bounds for optimisation
Plow = min(Pd)
Phigh = max(Pd)*4
Qlow = min(Qd)#*-20
Qhigh = max(Qd)#*20

#define step for discretization
size = 2; ideal = 1/size;
#define discrete step for each generator
source = [list(np.linspace(fva.iloc[i]['minimum'],fva.iloc[i]['maximum'],size)) for i in range(len(fva))]
source = [[round(k,2) for i,k in enumerate(source[i])] for i in range(len(source))]



#number of iterations
iterations = 10
time_k1 = []


for k in range(iterations):
    #All Combinations
    a_combinations = list(map(list,itertools.product([0, 1], repeat=len(fva))))
    #define list to track discrete-samples generated
    tracker = []
    #keeps track of unique samples generated
    samples = []
    #number of samples
    n = int(2**len(fva) / 4);  true_s = 0; infeasible_count = 0;    
    
    s_time = time.time()
    for i in range(n):
        t = a_combinations.pop()
        new_sample =  _generate_sample_1(t)
        if new_sample is not None:
            
            tracker.append(t)
            samples.append(new_sample)
#             a_combinations.remove(a_combinations[i])
            true_s += 1
        else:
            infeasible_count += 1
    
    #Subsequent loop
    gen_samples = size**len(fva)
    max_tries = None; try_ = 0;
    #Define weight parameters
    w_0 = 100
    base_val = 10
    #Define same-targets tracker
    num_targets = 0
    repetitions = 0
    while True:
        if max_tries is not None and try_ >= max_tries:
            break
        try_ += 1    

        #Update weights
        w = sum(np.array(tracker))/len(tracker)
        weights = w_0*abs(w-ideal) + base_val

        #Define targets
        targets = (np.round(w)+1)%2
        for i in range(len(w)):
            if w[i] == ideal:
                targets[i] = round(random.random())

        # If the targets have been sampled before? leave the smallest change
        if list(targets) in tracker:
            repetitions += 1
            targets = a_combinations[-1]


    #     print(targets)
        #Get actual targets
        a_targets = [source[i][int(targets[i])] for i in range(len(fva))]

        #Generate new sample
        new_sample =  _generate_sample_2(a_targets, weights)

        if new_sample is not None:
            r_target = [source[i].index(new_sample[i]) for i in range(len(fva))]
    #         print(r_target)
    #         print("\n")
            tracker.append(r_target)
            samples.append(new_sample)

            if r_target in a_combinations:
                a_combinations.remove(r_target)
            true_s += 1
        else:
            infeasible_count += 1 
            # change trajectory?
            tracker.append([round(random.random()) for i in range(len(fva))])
            print("Infeasible")
        if len([list(x) for x in set(tuple(x) for x in samples)]) == gen_samples: break
    
    
    time_k1.append(time.time() - s_time) 
```
